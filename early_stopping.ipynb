```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X,y = make_moons(n_samples=100, noise =0.2, random_state=1)
```

```python
X,y
```

```python
X.shape
```

```python
plt.scatter(X[:,0],X[:,1], c = y)
```

```python
X.shape
```

```python
y
```

```python
xtrain,xtest = X[:80],X[80:] #first 80 rows of X used for training , last 20 rows of X used for testing
ytrain,ytest = y[:80],y[80:] #first 80 rows of y used for training , last 20 rows of y used for testing
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtest.shape
```

```python
ytest.shape
```

```python
model = Sequential() #brain without neurons
```

```python
#input layer
model.add(Dense(500,activation = "relu", input_shape = (2,)))
```

```python
#output layer
model.add(Dense(1,activation="sigmoid"))
```

```python
#compile function
model.compile(optimizer="adam"  , loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
history = model.fit(xtrain,ytrain, epochs = 1000, validation_data = (xtest,ytest))
```

```python
plt.plot(history.history['loss'], label ="Training error")
plt.legend()
```

```python
plt.plot(history.history['val_loss'], label ="Testing Error")
plt.legend()
```

```python
plt.plot(history.history['loss'], label ="training error")
plt.plot(history.history['val_loss'], label ="testing error")
plt.legend()
```

# early stopping technique without patience

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
es = EarlyStopping(monitor="val_loss")
```

```python
model2 = Sequential()
model2.add(Dense(500,activation = "relu", input_shape = (2,)))
model2.add(Dense(1,activation="sigmoid"))
model2.compile(optimizer="adam"  , loss = "binary_crossentropy", metrics = ["accuracy"])
history = model2.fit(xtrain,ytrain, epochs = 1000, validation_data = (xtest,ytest), callbacks=[es])
```

```python
plt.plot(history.history['loss'], label ="training error")
plt.plot(history.history['val_loss'], label ="testing error")
plt.legend()
plt.legend()
```

# early stopping technique with patience

```python
from tensorflow.keras.callbacks import EarlyStopping
es = EarlyStopping(monitor="val_loss", patience=50)
model3 = Sequential()
model3.add(Dense(500,activation = "relu", input_shape = (2,)))
model3.add(Dense(1,activation="sigmoid"))
model3.compile(optimizer="adam"  , loss = "binary_crossentropy", metrics = ["accuracy"])
metadata = model3.fit(xtrain,ytrain, epochs = 1000, validation_data = (xtest,ytest), callbacks=[es])
```

```python
plt.plot(metadata.history['loss'], label ="training error")
plt.plot(metadata.history['val_loss'], label ="testing error")
plt.legend()
plt.legend()
```
